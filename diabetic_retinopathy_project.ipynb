```
from sklearn.preprocessing import StandardScaler
```

```
scaler = StandardScaler()
```

```
scaler.fit(df.drop('class_label',axis=1))
```

```
scaled_features = scaler.fit_transform(df.drop('class_label',axis=1))
```

```
df_feat = pd.DataFrame(scaled_features,columns=df.columns[:-1])
df_feat.head()
```

```
X = df_feat
```

```
y = df['class_label']
```

```
from sklearn.model_selection import train_test_split
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

# TENSORFLOW

```
import tensorflow as tf
```

```
X.columns
```

```
feat_cols = []
for col in X.columns:
    feat_cols.append(tf.feature_column.numeric_column(col))
```

```
feat_cols
```

```
input_func = tf.estimator.inputs.pandas_input_fn(x=X_train,batch_size=10,num_epochs=36905,shuffle=True)
```

```
classifier = tf.estimator.DNNClassifier(hidden_units = [20,10,20,10],n_classes=3,feature_columns=feat_cols)
```

```
classifier.train(input_fn=input_func,steps=50)
```

```
pred_fn = tf.estimator.inputs.pandas_input_fn(x=X_test,batch_size=len(X_test),shuffle=False)
```

```
predictions = list(classifier.predict(input_fn=pred_fn))
```

```
predictions
```

```
final_preds = []

for pred in predictions:
  final_preds.append(pred['class_ids'][0])
```

```
final_preds
```

```
from sklearn.metrics import classification_report,confusion_matrix
```

```
print(confusion_matrix(y_test,final_preds))
```

```
print(classification_report(y_test,final_preds))
```

# COMPARING WITH RANDOM FOREST

```
from sklearn.ensemble import RandomForestClassifier
```

```
rfc = RandomForestClassifier(n_estimators=200)
```

```
rfc.fit(X_train,y_train)
```

```
rfc_preds = rfc.predict(X_test)
```

```
print(classification_report(y_test,rfc_preds))
```

```
print(confusion_matrix(y_test,rfc_preds))
```
